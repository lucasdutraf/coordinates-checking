## Bibliotecas utilizadas

* **Shapely**: Manipulação e análise de objetos geométricos no plano cartesiano. https://shapely.readthedocs.io/en/stable/project.html
* **Geopy**: client para vários serviços web de geocodificação populares. https://geopy.readthedocs.io/en/stable/
* **Matplotlib**: Criar visualizações estáticas, animadas e interativas em Python. (Opcional)

```python
import matplotlib.pyplot as plt
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
```

Coordenadas de 4 capitais diferentes, utilizando o site https://www.latlong.net/

> As coordenadas precisam estar ordenadas em sentido horário. (Primeiramente o inferior-esquerdo, superior-esquerdo...)

```python
CITIES = {
    "Sao Paulo": [(-23.564931, -46.654846), (-23.540383, -46.655533), (-23.538652, -46.620008), (-23.563515, -46.617088)],
    "Rio de Janeiro": [(-22.944799, -43.276982), (-22.879655, -43.284196), (-22.879655, -43.186650), (-22.927725, -43.176687)],
    "Fortaleza": [(-3.773819, -38.571732), (-3.693652, -38.583528),(-3.686799, -38.472904), (-3.766282, -38.457101) ],
    "Distrito Federal": [(-16.061650, -48.270701), (-15.511912, -48.204739), (-15.506619, -47.435127), (-16.061650, -47.325190)],
}
```

```python
for city in CITIES.keys():
    polygon = Polygon(CITIES[city])
    print(f"\n{city}")
    display(polygon)
```

```python
print("São Paulo")
polygon = Polygon(CITIES["Sao Paulo"])
plt.plot(*polygon.exterior.xy)
```

```python
print("Rio de Janeiro")
polygon = Polygon(CITIES["Rio de Janeiro"])
plt.plot(*polygon.exterior.xy)
```

```python
print("Fortaleza")
polygon = Polygon(CITIES["Fortaleza"])
plt.plot(*polygon.exterior.xy)
```

```python
print("Distrito Federal")
polygon = Polygon(CITIES["Distrito Federal"])
plt.plot(*polygon.exterior.xy)
```
